```python
file='poco_data.csv'
with open (file,'w+') as f:
    writer=csv.writer(f)
    writer.writerow(['Ratings','Review'])
    
    for i in range(31):        
        link=f'https://www.flipkart.com/poco-x4-pro-5g-laser-blue-128-gb/product-reviews/itmeb324bd268ef4?pid=MOBGCGZZDHZ2HJU8&lid=LSTMOBGCGZZDHZ2HJU8ONUSOK&marketplace=FLIPKART&page={i}'
        
        page=requests.get(link)
        soup=bs(page.content,'html.parser')
        main_box=soup.find_all('div',{'class':'col _2wzgFH K0kLPL'})
        for box in main_box:
            try:
                
                rat=box.find('div',{'class':'_3LWZlK _1BLPMq'}).text.strip()
                rev=box.find('div',{'class':'t-ZTKy'}).text.strip()
                writer.writerow([rat,rev])
            except:AttributeError
f.close()
```
